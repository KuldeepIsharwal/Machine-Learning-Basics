#Problem with perceptron trick :

 - we used to correct the line multiple line to get line which seperates our classes well, but if seperation between classes it a bit more it seperates class but not right down the middle so it might perform poor on real data

- in perceptron trick misclassified points used to pull the line towards itself if the line was on the other side

- but now in this algo the correctly classified points will puch the line away from themselves ,and misclassified points will pull the line towards themselves, and the magnitude of this push/pull will be depend on how much far away the line is from the points

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=30)
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):

    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1

    for i in range(1000):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]

    return weights[0],weights[1:]

def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

#Maximum Likelihood Estimation (MLE)

A method of estimating the parameters of a statistical model by finding the values that maximise the probability (likelihood) of having observed the given training data — under the assumed model.



- Given data X = {x₁, x₂, ..., xₙ} and model parameters θ:


  Likelihood:   L(θ) = P(X | θ) = P(x₁|θ) × P(x₂|θ) × ... × P(xₙ|θ)


   Assumes observations are independent — so joint probability is the product of individual probabilities.




   
   Goal: find θ that maximises L(θ).

- Log-Likelihood — why we use it:
Products of many small probabilities underflow to zero numerically. Taking the log converts products to sums:


log L(θ) = -1 * (log P(x₁|θ) + log P(x₂|θ) + ... + log P(xₙ|θ))


Since log is monotonically increasing, maximising log L(θ) is identical to maximising L(θ).
