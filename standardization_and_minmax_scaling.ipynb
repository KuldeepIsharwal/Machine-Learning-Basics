```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
df.head()
```

```python
df = df.iloc[:,2 :]
df.sample(5)
```

Train Test split :

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('Purchased',axis =1), df['Purchased'],test_size = 0.3,random_state = 0)
```

Standard Scaler :

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(x_train)
x_train = scaler.transform(x_train)
# here we only fit x_train but we transform both x_train and x_test
x_test = scaler.transform(x_test)
```

```python
scaler.mean_
```

```python
x_train[:5]
```

```python
#Since our standard scaler has changed our pandas dataframe to numpy array
#We need to convert it back to pandas dataframe
x_train_scaled = pd.DataFrame(x_train, columns = df.drop('Purchased', axis=1).columns)
x_test_scaled = pd.DataFrame(x_test, columns = df.drop('Purchased', axis=1).columns)
```

Effect of scaling

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr_scaled = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
lr_scaled.fit(x_train_scaled,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred_scaled = lr_scaled.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score
print("actual",accuracy_score(y_test,y_pred))
print("scaled",accuracy_score(y_test,y_pred_scaled))
```
