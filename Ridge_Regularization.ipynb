```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
```

#Regularisation

A set of techniques that reduce overfitting by adding a penalty term to the loss function that discourages the model from learning overly complex patterns — forcing it to stay simpler and generalise better to unseen data.



- It is of three types :
   

1.    Ridge Regression (L2) : Penalty = sum of     squared weights.


     Loss = MSE + λ × Σwᵢ²


- Built-in automatic feature selection

-  Produces sparse models — easier to interpret

- Unstable when features are correlated — arbitrarily picks one and zeros the rest


2.    Lasso Regression (L1)


3.    ElasticNet (L1 + L2 combined)



image.png

```python
x,y = make_regression(n_samples=100, n_features=1,n_informative=1,n_targets = 1, noise=20, random_state = 13)
```

```python
plt.scatter(x,y)
```

```python
from sklearn.linear_model import LinearRegression,Ridge
```

```python
lr = LinearRegression()
```

```python
lr.fit(x,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
rr = Ridge(alpha=10)
rr.fit(x,y)
print(rr.coef_)
print(rr.intercept_)
```

```python
rr1 = Ridge(alpha=100)
rr1.fit(x,y)
print(rr1.coef_)
print(rr1.intercept_)
```

```python
plt.plot(x,y,'b.')
plt.plot(x,lr.predict(x),'r.')
plt.plot(x,rr.predict(x),'g.')
plt.plot(x,rr1.predict(x),'y.')
```
