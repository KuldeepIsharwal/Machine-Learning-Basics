```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np

import plotly.express as px
import plotly.graph_objects as go

from sklearn.metrics import mean_squared_error, r2_score
```

```python
x,y = make_regression(n_samples=100,n_features=2,n_informative=2,n_targets=1,noise =50)
```

```python
df = pd.DataFrame({'feature1': x[:,0],'feature2': x[:,1],'target': y})
df.head()
```

```python
fig = px.scatter_3d(df,x='feature1',y='feature2',z='target')
fig.show()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
X = np.linspace(-5, 5, 10)
Y = np.linspace(-5, 5, 10)
xGrid, yGrid = np.meshgrid(X, Y)

# Define final before it is used
final = np.vstack((xGrid.ravel(),yGrid.ravel())).T

z_final = lr.predict(final).reshape(10,10)

Z = z_final
```

```python
fig = px.scatter_3d(df, x='feature1', y='feature2', z='target')

fig.add_trace(go.Surface(x = X, y = Y, z = Z))

fig.show()
```

#Assumptions of LR :



1.   Linear Relationship between input and output


2.   No multi collinearity (check : clac. variance_inflation_factor OR calc correlation)


3. Normality of Residual : the  residual (Ypred -Yactual) should be normal


4.  No autocorrelation of error : when u plot your residual there should be no pattern (any known curve) it should be zigzag
