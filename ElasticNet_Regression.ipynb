#ElasticNet (L1 + L2 combined)
Penalty = mix of both L1 and L2.
Loss = MSE + λ × [l1_ratio × Σ|wᵢ| + (1 - l1_ratio) × Σwᵢ²]

-  When we have really large dataset > 100 features we apply Elastic Net

- When cols have multi-collinearity we should use this

```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression,Ridge,Lasso,ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(load_diabetes().data,load_diabetes().target,test_size=0.2,random_state=2)
```

```python
#Linear Regression
reg = LinearRegression()
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
#Ridge
reg = Ridge(alpha=0.1)
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
#Lasso
reg = Lasso(alpha=0.01)
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
#ElasticNet
reg = ElasticNet(alpha=0.005,l1_ratio=0.9)
reg.fit(x_train,y_train)
y_pred = reg.predict(x_test)
r2_score(y_test,y_pred)
#
```
