#Feature Construction

The process of creating entirely new features from existing raw data — using domain knowledge, mathematical operations, or interactions between variables — that better represent the underlying structure of the problem for a machine learning model.



```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/titanic.csv',usecols=['Pclass','Age','SibSp','Parch','Survived'])
```

```python
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

#Applying Feature construction :
  new coln named family size = sibsp+parch

```python
df['Family Size'] = df['Parch'] + df['SibSp']+1
```

```python
df.sample(5)
```

```python
df.drop(columns=['SibSp','Parch'],inplace=True)
```

```python
df.sample(5)
```
