#Logistic Regression

A supervised classification algorithm that models the probability that an input belongs to a particular class — using the logistic (sigmoid) function to squash a linear combination of features into a value between 0 and 1.

#Perceptron Trick :
1. Start with a random line.

2. Pick a random data point from your dataset.

3. Check if the line classified it correctly.

4.1 If it's correct: Do nothing.

4.2 If it's misclassified: Change the line's position by moving it closer to/over the misclassified point.

```python
from sklearn.datasets import make_classification
import numpy as np
x,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41, hypercube = False,class_sep =10)
print(x.shape,y.shape)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap = 'winter',s=100)
plt.show()
```

```python
def perceptron(x,y) :
  x = np.insert(x,0,1,axis=1)
  weights = np.ones(x.shape[1])
  lr = 0.1

  for i in range(1000):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(x[j],weights))
    weights = weights - lr * (y[j] - y_hat)*x[j]

  return weights[0],weights[1,:]
```

```python
def step(z):
  return 1 if z>0 else 0
```
