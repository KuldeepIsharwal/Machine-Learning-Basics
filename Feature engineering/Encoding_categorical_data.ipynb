#COLUMN TRANSFORMER
Real datasets have mixed column types — some numerical, some nominal categorical, some ordinal. Each needs a different treatment. Doing this manually means writing separate code for each column and carefully concatenating results back together. ColumnTransformer automates this into one clean object.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

ct = ColumnTransformer(transformers=[
    ('onehot', OneHotEncoder(drop='first', sparse_output=False), ['colour']),
    ('ordinal', OrdinalEncoder(categories=[['small','medium','large']]), ['size'])
], remainder='passthrough')

X_transformed = ct.fit_transform(df)
```
