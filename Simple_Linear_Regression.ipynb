#Linear Regression

A supervised learning algorithm that models the relationship between one or more input features and a continuous numerical output by fitting a straight line (or hyperplane) that minimises the sum of squared differences between predicted and actual values.

General en for simple lr is - y = mx+b

lr regression calculates m and b for us

#Types :

1.   Simple Linear Regression
2.   Multiple Linear Regression
3.   Polynomial Linear Regression



# two ways to calculate m and b


1.   Closed form (OLS)

image.png


2.   Non - closed form (Gradient Descent)



#Regression Metrics :    



1.   MAE — Mean Absolute Error

Average of absolute differences. Easy to interpret — same units as target.

Adv - unit same as Y
      Robust to outliars


Disadv -  Not differentiable at zero so cannot find absolute low by putting d/d(m,b) = 0


2.   MSE — Mean Squared Error

Average of squared differences. Penalises large errors heavily.

adv - it is differentiable

disadv - not robust to outliars


3.   RMSE — Root Mean Squared Error

Square root of MSE — back in original units.

disadv - not robust to outliars

4.   R² Score (coefficient of determination)

Fraction of variance in y explained by the model. Range: 0 to 1 (can be negative for terrible models).



```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/placement (1).csv')
```

```python
df.sample(5)
```

```python
from matplotlib import pyplot as plt

plt.scatter(df['cgpa'],df['package'])
plt.show()
```

```python
x = df.iloc[: , 0:1]
y = df.iloc[: , -1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score,root_mean_squared_error
```

```python
y_pred = lr.predict(x_test)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
root_mean_squared_error(y_test,y_pred)
```
