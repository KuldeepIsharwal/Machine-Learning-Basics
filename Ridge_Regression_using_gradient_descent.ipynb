```python
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import numpy as np
```

```python
x,y = load_diabetes(return_X_y=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
reg = SGDRegressor(penalty='l2',max_iter=100,eta0=0.1,learning_rate='constant',shuffle=True)
```

```python
reg.fit(x_train,y_train)

y_pred = reg.predict(x_test)
r2_score(y_test,y_pred)
```

#key points of ridge regression :

- with increase in value of lembda the value of weights decreases but NEVER ZERO


- if value of coeff is really high it shrinks faster than coeff with relatively small value cz lemda acts on square of coeff

- Bias variance tradeoff : regularization with small lemda we get low B and high V and if we've high lemda we get high B and low V

- 'Ridge' naam kyun hai? : Iska naam isliye 'Ridge' hai kyunki iska mathematical solution hamesha ek circle (ya sphere) ki boundary/perimeter par lie karta hai, jise geometric view mein dekha ja sakta hai.
